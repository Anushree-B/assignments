## multi indexing, grouping and agregation

```python
import pandas as pd
import numpy as np
```

```python
#reading the dataset
df = pd.read_csv("Z:/python sem 2/stocks.csv")
```

```python
df
```

```python
# set index on 'Symbol' and 'Date'
df.set_index(['Symbol','Date'],inplace=True)
df
```

```python
# sort the index
df.sort_index()
```

```python
# select data for AAPL
df.loc['AAPL']
```

```python
#select "close" value for "AAPL" on "2016-10-03"
df.loc[[('AAPL','2016-10-03')],['Close']]
```

```python
#select data for apple and microsoft
df.loc[['AAPL','MSFT']]
```

```python
#select data for apple on multiple dates
df.loc[[('AAPL','2016-10-03'), ('AAPL','2016-10-05')]]
```

```python
#selecting all data for 2 dates
df.loc[(slice(None), ['2016-10-03', '2016-10-04']),:]
```

```python
#fetch all rows with volume greater than 1800000
df.query("Volume>18000000")
```
